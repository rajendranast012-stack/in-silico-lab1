```python
import urllib.request
from urllib.parse import quote
from urllib.error import HTTPError, URLError

def CIRconvert(identifier, timeout=15):
    url = f"https://cactus.nci.nih.gov/chemical/structure/{quote(identifier)}/smiles"
    try:
        with urllib.request.urlopen(url, timeout=timeout) as response:
            smiles = response.read().decode("utf-8").strip()
            return smiles if smiles else None
    except HTTPError:
        return None
    except URLError:
        return None
    except Exception:
        return None

identifiers = ['Aspirin', '3-Methylheptane', 'Quercetin']
for name in identifiers:
    print(name, ":", CIRconvert(name))
```

```python
import urllib.request
from urllib.parse import quote
from urllib.error import HTTPError, URLError

def CIRconvert(identifier, timeout=15):
    if not isinstance(identifier, str) or identifier.strip() == "":
        return None

    url = f"https://cactus.nci.nih.gov/chemical/structure/{quote(identifier)}/smiles"
    try:
        with urllib.request.urlopen(url, timeout=timeout) as response:
            smiles = response.read().decode("utf-8").strip()
            return smiles if smiles else None
    except (HTTPError, URLError, Exception):
        return None
```
